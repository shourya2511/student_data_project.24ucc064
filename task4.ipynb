```python
df['age_standardized'] = (df['age'] - df['age'].mean()) / df['age'].std()
```

```python
df['studytime_standardized'] = (df['studytime'] - df['studytime'].mean()) / df['studytime'].std()
```

```python
np.std(df['studytime_standardized'])
```

```python
 df['GPA_standardized'] = (df['GPA'] - df['GPA'].mean()) / df['GPA'].std()
```

```python
df['absences_standardized'] = (df['absences'] - df['absences'].mean()) / df['absences'].std()
```

```python
df = pd.get_dummies(df, columns=['sex'], drop_first=True)


```
