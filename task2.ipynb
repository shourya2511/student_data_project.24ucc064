```python
def get_grade(marks):
    if marks>=0 and marks<=3:
        return "F"
    elif marks>3 and marks <=6:
        return "E"

    elif marks>6 and marks <=9:
        return "D"

    elif marks>9 and marks <=12:
        return "C" 

    elif marks>12 and marks <=15:
        return "B"

    elif marks>15 and marks <=18:
        return "AB"   

    elif marks>18 and marks <=20:
        return "A" 

    else:
        return "Invalid Grade"


```

```python
df['grade'] = df['G_final'].apply(get_grade)
```

```python
df
```

```python
grade_to_num = {'A': 10, 'AB': 8, 'B': 7, 'C': 6, 'D':5,'E':4,'F':2,}
```

```python
df['Grade_Numeric'] = df['grade'].map(grade_to_num)
```
